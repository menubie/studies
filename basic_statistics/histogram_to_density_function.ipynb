## Perguntas

- Qual o tipo de cálculo feito para chegarmos em uma função densidade a partir de um histograma?
- No gráfico de densidade, o eixo y está em qual grandeza?

a) Histograma de uma variável de comportamento normal e função densidade da mesma variável

```python
t = pd.DataFrame(np.random.normal(size=10000, loc=50), columns=['beta'])

t.plot.hist(bins=100, color='lightcoral')

t.plot.hist(bins=100, color='lightcoral', cumulative=True)

t.plot.density(color='lightcoral')
```

b) Histograma de uma variável de comportamento completamente _random_ e função densidade da mesma variável

```python
v = pd.DataFrame(np.random.random(size=10000), columns=['random_'])

v.plot.hist(bins=100, color='gold')

v.plot.hist(bins=100, color='gold', cumulative=True)

v.plot.density(color='gold')
```
